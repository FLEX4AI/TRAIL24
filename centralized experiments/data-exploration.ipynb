# This notebook show a bit of data exploration on one of the buildings. test_buildings dataset contains the first 9 buildings of the original data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
file_path = '/kaggle/input/test-buildings/residential_1005.pkl'

with open(file_path, 'rb') as file:
    data = pickle.load(file)

print(data)
```

```python
#selecting only the consumption and date time columns
data=data[["consumption","date_time"]]
data.set_index(['date_time'], inplace=True)
print(data)
```

```python
#plotting the data (30mins time span)
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['consumption'], label='Energy Consumption')
plt.title('Energy Consumption Over Time')
plt.xlabel('Date Time')
plt.ylabel('Consumption')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#plotting the data (1h time span)
data_hourly = data.resample('h').sum()
plt.figure(figsize=(14, 7))
plt.plot(data_hourly.index, data_hourly['consumption'], label='Energy Consumption')
plt.title('Hourly Energy Consumption Over Time')
plt.xlabel('Date Time')
plt.ylabel('Consumption')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#plotting the data (1 day time span)
data_daily = data.resample('D').sum()
plt.figure(figsize=(14, 7))
plt.plot(data_daily.index, data_daily['consumption'], label='Energy Consumption')
plt.title('Daily Energy Consumption Over Time')
plt.xlabel('Date Time')
plt.ylabel('Consumption')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#plotting the data (1 week time span)
data_weekly = data.resample('W').sum()
plt.figure(figsize=(14, 7))
plt.plot(data_weekly.index, data_weekly['consumption'], label='Energy Consumption')
plt.title('Weekly Energy Consumption Over Time')
plt.xlabel('Date Time')
plt.ylabel('Consumption')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#plotting the data (1 month time span)
data_monthly = data.resample('ME').sum()
plt.figure(figsize=(14, 7))
plt.plot(data_monthly.index, data_monthly['consumption'], label='Energy Consumption')
plt.title('Monthly Energy Consumption Over Time')
plt.xlabel('Date Time')
plt.ylabel('Consumption')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#plotting the data (1 year time span)
data_yearly = data.resample('YE').sum()
plt.figure(figsize=(14, 7))
plt.plot(data_yearly.index, data_yearly['consumption'], label='Energy Consumption')
plt.title('Yearly Energy Consumption Over Time')
plt.xlabel('Date Time')
plt.ylabel('Consumption')
plt.legend()
plt.grid(True)
plt.show()
```
